```python
import pandas as pd
```

```python
data = pd.read_excel('data/data_rakamin_customers.xlsx')
```

```python
data.info()
```

**statistics type**

**count** menghitung total baris

**nunique** menghitung total baris yang unique

**mean** rata-rata dari kolom

**median** median data dari kolom

**min** nilai terkecil dari kolom

**max** nilai maksimal dari kolom


```python
#Ada berapa customer pada suatu provinsi
data.groupby('province').agg({'registration_id' : 'count'})
```

```python
#Ada berapa unique customer pada suatu provinsi
data.groupby('province').agg({'registration_id' : 'nunique'})
```

```python
#Ada berapa rata-rata umur pada suatu provinsi
data.groupby('province').agg({'age' : 'mean'})
```

```python
#Ada berapa rata-rata umur per income_group
data.groupby('income_group').agg({'age' : 'mean'})
```

```python
#Berapa median website_visit_count per income_group
data.groupby('income_group').agg({'website_visit_count' : 'median'})
```

```python
#Berapa rata-rata , median , termuda dan tertua umur per province?
data.groupby('province').agg({'website_visit_count' : ['mean','median','min','max']})
```

```python
#Berapa rata-rata , dan maksimum age per website_cs_call_count
data.groupby('website_cs_call_count').agg({'age' : ['mean','max']})
```

```python
#Hitunglah banyaknya customer, rata-rata & median umur, dan
#website_purchase_rate minimum & maximum per income_group?
data.groupby('income_group').agg({'registration_id' : ['count'], 'age' : ['mean','median'], 'website_purchase_rate' : ['min','max']})
```

```python
# Hitunglah median website_recency, rata-rata website_purchase_amount,
# dan rata-rata website_visit_count per city?
data.groupby('city').agg({'website_recency' : ['median'], 'website_purchase_amount' : ['mean'], 'website_visit_count' : ['mean']})
```

```python
# Hitunglah banyaknya customer, rata-rata & median umur, dan
# website_purchase_rate minimum & maximum per province dan
# income_group?

data.groupby(['province','income_group']).agg({'registration_id' : ['nunique'],
                                               'age' : ['mean','median'],
                                               'website_purchase_rate' : ['min','max']})
```

```python
# Hitunglah median website_recency, rata-rata website_purchase_amount,
# dan rata-rata website_visit_count per province, city dan income_group?

data.groupby(['province','city']).agg({'registration_id' : ['nunique'],
                                               'age' : ['mean','median'],
                                               'website_purchase_rate' : ['min','max']})
```

```python
# Hitunglah median website_recency, rata-rata website_purchase_amount,
# dan rata-rata website_visit_count per province, city dan income_group?

data.groupby(['province','city','income_group']).agg({'website_recency' : ['median'],
                                                      'website_purchase_amount' : ['mean'],
                                                      'website_visit_count' : ['mean']})
```

```python
# data_res = data.reset_index()
data_res = data.reset_index(drop=True) #Supaya setelah reset index tidak menjadi kolom di dataframe
data_res
```

```python
#mereset hasil describe
df_group_by_reset = data.groupby(['province','city','income_group']).agg({'website_recency' : ['median'],
                                                      'website_purchase_amount' : ['mean'],
                                                      'website_visit_count' : ['mean']})
df_group_by_reset = df_group_by_reset.reset_index()
df_group_by_reset
```

```python
#How to rename column name
df_customers = data.rename(columns = {'age' : 'usia_customer'}) #kalau ingin merubah banyak column tinggal pisahkan saja dengan comma dengan format json
df_customers
```

```python
pd.pivot_table(df_customers,
               index=['income_group'],
               values=['registration_id','usia_customer','website_purchase_rate'],
               aggfunc={'registration_id' : ['nunique'],
                        'usia_customer' : ['mean','median'],
                        'website_purchase_rate' : ['min', 'max']
                        })
```
