**String Conversion**

Series.str.lower
Converts all characters to lowercase.

Series.str.upper
Converts all characters to uppercase.

Series.str.title
Converts first character of each word to uppercase and remaining to lowercase.

Series.str.capitalize
Converts first character to uppercase and remaining to lowercase.

Series.str.swapcase
Converts uppercase to lowercase and lowercase to uppercase.

Series.str.casefold
Removes all case distinctions in the string.

**Pattern**

Pattern DIMULAI dengan: df[ df[â€˜column_nameâ€™].str.contains(â€˜^patternâ€™) ]
Pattern DIAKHIRI dengan: df[ df[â€˜column_nameâ€™].str.contains(â€˜pattern$â€™) ]

```python
#filter awalan dan akhiran mengandung sesuatu
#mencari data dengan kolom nama berawalan "a"
new_filter = data[data['nama'].str.contains('^a')]
new_filter
```

```python
#filter awalan dan akhiran mengandung sesuatu
#mencari data dengan kolom telepon berawalan "62853"
new_filter = data[data['telepon'].astype('str').str.contains('^62853')]
new_filter
```
