```python
def resize_img(img) :
    original_width, original_height = img.size
    new_width = int(original_width * 0.5)
    new_height = int(original_height * 0.5)
    
    return {
        'new_width' : new_width,
        'new_height' : new_height
    }
```

```python
#Rubah Gambar Disini
image1 = 'https://foto.kontan.co.id/StXCKE_cdjxl3ync1Wq9iTMX4zg=/smart/2023/03/09/1036754397p.jpg'
image2 = 'https://img.okezone.com/content/2023/08/13/33/2863337/gara-gara-ini-raffi-ahmad-alami-masalah-otot-WHPlpedhhb.jpg' 
# image2 = 'https://asset.kompas.com/crops/m5XeaXLOI5PUYiSr5XpEy2EEdL8=/0x120:750x620/750x500/data/photo/2022/11/15/637327d9a47dc.jpg' #Foto saat menggunakan kacamata

path1 = 'photos/uploaded/'+str(uuid.uuid4())+'.'+(image1.split('.')[-1])
path2 = 'photos/compare/'+str(uuid.uuid4())+'.'+(image2.split('.')[-1])

General.save_image_from_url(image1, path1)
General.save_image_from_url(image2, path2)

print()

#Display and Resize Image
img1 = Image.open(path1)
img2 = Image.open(path2)
resize1 = resize_img(img1)
resize2 = resize_img(img2)

print(resize1)
display(img1.resize((resize1['new_width'], resize1['new_height'])))
display(img2.resize((resize2['new_width'], resize2['new_height'])))

#Face Comparison
result = DeepFace.verify(img1_path = path1, img2_path = path2)

if(result['verified']) :
    print('###### Face match #####')
else:
    print('###### Face Not Match ######')
    
os.remove(path1)
os.remove(path2)

print()
print("Detail :")
print(result)
```
