## Iterasi Dictionary

### Memanggil key pada Dictionary

```python
seller_kategori = {'sellerA':['Fashion'],
                  'sellerB':['Fashion', 'Elektronik'],
                  'sellerC':['Perabotan', 'Buku'],
                  'sellerD':['Elektronik'],
                  'sellerE':['Fashion'],
                  'sellerF':['Perabotan', 'Elektronik']}

for i in seller_kategori:
    print(i) #key pada dictionary
```

### Memanggil value pada Dictionary

```python
for i in seller_kategori:
    print(seller_kategori[i])
```

### Memanggil key dan value sekaligus pada Dictionary

```python
# memanggil key dan value sekaligus
for i in seller_kategori.items():
    print(i) #berupa tuple
```

```python
# memanggil key dan value sekaligus
for key, value in seller_kategori.items():
    print('Nilai Key', key, 'dan value adalah', value) #berupa tuple
```

### Membuat Dictionary Baru pada For Loop

```python
new_dict = {} #membuat dict kosong

for key, value in seller_kategori.items():
    if 'Perabotan' in value: #check
        new_dict.update({key:value}) #menggunakan fungsi update

print(new_dict)
```

### Mini Quiz | Challenge Time #4

Gunakan dataset <b>monthly_rakamin_customer_order!</b> Team bisnis memiliki rule, jika dalam sebulan

- Total order bulanan customer > rata-rata total order bulanan semua customer di bulan tersebut (mean Order customer monthly)
- Total GMV bulanan customer > rata-rata total GMV bulanan semua customer di bulan tersebut (mean GMV customer monthly)

Maka keputusan bisnisnya user eligible untuk diberikan voucher discount. Ada berapa user yang menjadi user eligible pada bulan Maret 2022?

<b>Hint: </b> Buat kolom mean_total_order_monthly dan mean_total_amount_monthly (Bisa dengan cara transform), kemudian lakukan iterrows

```python
#Challenge Time #4

import pandas as pd
import numpy as np

pd.options.display.float_format = '{:,.0f}'.format #formating float
```

```python
df = pd.read_csv('monthly_rakamin_customer_order.csv')
df.head()
```

```python
# check = df.groupby(['user_id','order_month']).agg({'total_order' :'count'})
# check
```

```python
# Mencari Rata-rata total order dan total gmv
df_mean = df.groupby(['order_month']).agg({'total_order' : 'mean', 'total_gmv' : 'mean'}).reset_index()
df_mean.columns = ['order_month', 'mean_total_order', 'mean_total_gmv']
df_mean

#Menggabungkan Data
df = df.merge(df_mean, how='inner', on='order_month')
df.head()

#melakukan pengecekan eligible atau tidak
eligible = []
for index, column in df.iterrows():
  total_order = column['total_order']
  total_gmv = column['total_gmv']
  mean_order = column['mean_total_order']
  mean_gmv = column['mean_total_gmv']

  if (total_order > mean_order) and (total_gmv > mean_gmv) :
    eligible.append(True)
  else:
    eligible.append(False)

#membuat kolom baru dan menambahkan data is eligible
df['is_eligible'] = eligible

#filter data bualn maret yang eligible dan tidak
df[df['order_month'] == '2022-03-01'].groupby('is_eligible').agg({'user_id' : 'nunique'}).reset_index()
```

```python
# Terdapat 50 user dengan user_id berurut dari 101 - 150,
# kita akan memilih beberapa user_id saja yang akan dijadikan sebagai target user interview. Kriteria user yang akan diambil adalah.
# Jika nilai user_id adalah bilangan kelipatan 5 yang habis dibagi 2.
# Ada berapa jumlah user yang menjadi target participant?
# Buatlah dua versi, pertama menggunakan iterasi for dan kedua menggunakan while!
# Hint: Buatlah variabel dengan value 0 yang terus bertambah secara incremental setiap iterasi jika memenuhi kondisi yang diminta, gunakan conditional statement - I

#Logika For
eligible = 0
for i in range(101,151):
  if i%5 == 0 and i%2 == 0:
    eligible += 1

print(eligible)

# Logika While
start = 101
end = 150
eligible = 0

while start <= end:
  if start%5 == 0 and start%2 == 0:
    eligible += 1
  start +=1

print(eligible)
```

```python
from math import comb
from itertools import combinations

#Find Combination
list_id_anggota = ['a1', 'a2', 'a3', 'a4', 'a5', 'a6', 'a7', 'a8']
jumlah_anggota_tim = 2
jumlah_kemungkinan_tim = comb(len(list_id_anggota), jumlah_anggota_tim)
print(f"Jumlah Kombinasi: {jumlah_kemungkinan_tim} tim")

#Here is the combination
for tim in combinations(list_id_anggota, jumlah_anggota_tim):
    print(f"Tim: {tim}")
    jumlah_kemungkinan_tim += 1
```
